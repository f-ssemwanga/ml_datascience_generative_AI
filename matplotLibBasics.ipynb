#### Matplotlib Basics

#### Draw line Graphs

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-3, 3, 0.001) #create x axis with points from -3 to 3 with increments of 0.001

plt.plot(x, norm.pdf(x)) # y is the probability density function with normal distribution of x
plt.show()
```

#### Multiple plots on one graph

```python
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5)) #on the y we have a mean of 1.0 and std of 0.5
plt.show()
```

#### Saving plots to a file

```python
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5)) #on the y we have a mean of 1.0 and std of 0.5
plt.savefig('./resources/linePlots.png', format='png')
```
