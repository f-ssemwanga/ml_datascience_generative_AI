#### Seaborn

Seaborn is a visualisation library that sits on top of matplotlib, making it nicer to look at and adding some extra capabilities.

We are using a real dataset on 2019 model-year-vehicles, and plotting a histogram just using matplotlib to see the distribution by number of gears

```python
%matplotlib inline
import pandas as pd
import matplotlib as plt

df = pd.read_csv("./resources/fuelEfficiency.csv")
gear_counts = df['# Gears'].value_counts() #how many times each unique value appears

gear_counts.plot(kind="bar")
```

Loading up Seaborn and calling set() on to it to change matplotlib's default settings to something more visually pleasing


```python
import seaborn as sns

sns.set()

gear_counts.plot(kind="bar")
```

```python
#looking at the data closely
df.head()
```

Using the "distplot" to plot a histogram together with a smooth distribution of the histogram overlaid on it.

```python
#Distribution of MPG values on the vehicles
sns.displot(df['CombMPG'])
```

#### Pair plot

Let you visualise plots of every combination of various attribute together so you can look for interesting patterns between features.

In the example below we classify cars by how many cylinders are in the engines, and look for relationships between the cylinders, city MPG rating, Highway MPG rating and combined MPG rating.

```python
df2 = df[['Cylinders', 'CityMPG', 'HwyMPG', 'CombMPG']]
df2.head()
```

```python
sns.pairplot(df2, hue='Cylinders', height=2.5)
```

```python
sns.scatterplot(x="Eng Displ", y="CombMPG", data=df)
```

```python
#joint plot combines a scatterplot with histograms on both axes and lets you visualise both the individual data points and the distribution across both dimensions at the same time

sns.jointplot(x="Eng Displ", y="CombMPG", data=df)
```

#### lmplot

A scatter plot with a linear regression line applied to it and overlaid on to the data.

```python
sns.lmplot(x="Eng Displ", y="CombMPG", data=df)
```

#### box plot

Box and whiskers plot is used for visualising typical values for a given category without getting distracted by outliers

Each box represents the range between the first and 3rd quartile of the data with a line representing the median value.  THe "whiskers" extend from the box represent the spread of the remainder of the data, apart from the bear outliers that are plotted as individual  points outside of the whiskers.

The box plot below loks at box plots for each vehicle manufacturer, visualising the miles per gallon ratings across the vehicles they produce.  Tick labels will be rotated at 45 degrees for easy reading.



```python
sns.set(rc={'figure.figsize': (15,5)})
ax=sns.boxplot(x='Mfr Name', y='CombMPG', data=df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```

#### Swarm plot

Plots each individual data point but does so in such a way that groups them together based on their distribution.

```python
ax=sns.swarmplot(x='Mfr Name', y='CombMPG', data=df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```

#### Count plot 

For categorical dat - lets you count up how many times each given category on the X axis occurs in your data

```python
ax=sns.countplot(x='Mfr Name',  data=df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```

#### Heat Map

Allows you to plot tabular 2D data of some sort with colors representing  the individual values in each cell of the 2D table

The resulting heat map below shows all of the engine displacement values along the x axis and all of the cylinder values along the y-axis.

```python
df2 = df.pivot_table(index="Cylinders", columns="Eng Displ", values="CombMPG", aggfunc="mean") # aggregate values using a mean
sns.heatmap(df2)
```
