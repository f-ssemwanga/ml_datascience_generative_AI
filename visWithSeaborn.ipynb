#### Seaborn

Seaborn is a visualisation library that sits on top of matplotlib, making it nicer to look at and adding some extra capabilities.

We are using a real dataset on 2019 model-year-vehicles, and plotting a histogram just using matplotlib to see the distribution by number of gears

```python
%matplotlib inline
import pandas as pd
import matplotlib as plt

df = pd.read_csv("./resources/fuelEfficiency.csv")
gear_counts = df['# Gears'].value_counts() #how many times each unique value appears

gear_counts.plot(kind="bar")
```

Loading up Seaborn and calling set() on to it to change matplotlib's default settings to something more visually pleasing


```python
import seaborn as sns

sns.set()

gear_counts.plot(kind="bar")
```

```python
#looking at the data closely
df.head()
```

Using the "distplot" to plot a histogram together with a smooth distribution of the histogram overlaid on it.

```python
#Distribution of MPG values on the vehicles
sns.displot(df['CombMPG'])
```

#### Pair plot

Let you visualise plots of every combination of various attribute together so you can look for interesting patterns between features.

In the example below we classify cars by how many cylinders are in the engines, and look for relationships between the cylinders, city MPG rating, Highway MPG rating and combined MPG rating.

```python
df2 = df[['Cylinders', 'CityMPG', 'HwyMPG', 'CombMPG']]
df2.head()
```

```python
sns.pairplot(df2, hue='Cylinders', height=2.5)
```

```python
sns.scatterplot(x="Eng Displ", y="CombMPG", data=df)
```

```python
#joint plot combines a scatterplot with histograms on both axes and lets you visualise both the individual data points and the distribution across both dimensions at the same time

sns.jointplot(x="Eng Displ", y="CombMPG", data=df)
```
