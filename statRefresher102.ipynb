### Standard Deviation and Variance

#### Normal Distribution 

Means that values are concentrated around the mean

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

incomes = np.random.normal(1000, 50.0, 100000)
plt.hist(incomes,50)
plt.show()
```

```python
incomes.std()
```

```python
incomes.var()
```

### Data Distributions

#### Uniform Distribution

Means there is a flat constant probability of a value occurring within a given range

```python
values = np.random.uniform(-10.0, 10.0, 100000)
plt.hist(values, 50)
plt.show()
```

#### Normal /Gaussian

Visualises the probability density function


```python
from scipy.stats import norm

x = np.arange(-2,3,0.001)  #increment of 0.001
plt.plot(x, norm.pdf(x)) #y axis is normal probability density function

# on the graph below 0 represents the mean and the numbers at the bottom are the standard deviations
```

#### Generating some random numbers with a normal distribution

"mu" = desired mean to centre data around

"sigma" = standard deviation

```python
mu = 5.0
sigma = 2.0
values = np.random.normal(mu, sigma, 10000)
plt.hist(values, 50)
plt.show()
```

#### Exponential Probability Distribution Function / "Power Law"


```python
from scipy.stats import expon

x = np.arange(0, 10, 0.001)
plt.plot(x, expon.pdf(x))
```

#### Binomial probability mass function

```python
from scipy.stats import binom

n, p = 10, 0.5 # specify the shape of the data
x = np.arange(0, 10, 0.001)
plt.plot(x, binom.pmf(x, n, p))
```

#### Poisson probability Mass Function

Example: My website gets on average 500 visits per day, what is the odds of getting 550?

(Predicting number of things that happen in some given time period)

```python
from scipy.stats import poisson

mu = 500 #average
x = np.arange(400, 600, 0.5)
plt.plot(x, poisson.pmf(x,mu))
```
